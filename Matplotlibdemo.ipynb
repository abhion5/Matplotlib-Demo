# Bar plot

```python
companynames=['Google','Microsoft','Apple']
revenues=[345,210,267]
plt.xlabel("Company Names")
plt.ylabel("Revenue in Mn$")
plt.bar(companynames,revenues, color='m',linewidth=2,edgecolor='k')
```

```python
profit=[170,98,124]
plt.bar(companynames,profit)
```

```python
plt.bar(companynames,revenues,label="Revenue")
plt.bar(companynames,profit,label="Profit")
plt.legend()
```

```python
import numpy as np
x=np.arange(len(companynames))
x
```

```python
plt.ylabel("In Mn$")
plt.xticks(x,companynames)
plt.bar(x,revenues,width=0.4,label="Revenue",linewidth=2,edgecolor='k')
plt.bar(x+0.4,profit,width=0.4,label="Profit",linewidth=2,edgecolor='g')
plt.legend()
```
